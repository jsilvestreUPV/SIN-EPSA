# Ejercicio 2

Supón que estamos aplicando el algoritmo Perceptrón, con factor de aprendizaje $\alpha=1$ y margen $b=0.1$, a un conjunto de $4$ muestras bidimensionales de aprendizaje para un problema de $4$ clases, $c=1,2,3,4$. En un momento dado de la ejecución del algoritmo se han obtenido los vectores de pesos $\boldsymbol{w}_1=(-2,-2,-6)^t$, $\boldsymbol{w}_2=(-2,-2,-6)^t$, $\boldsymbol{w}_3=(-2,-4,-4)^t$, $\boldsymbol{w}_4=(-2,-4,-4)^t$. Suponiendo que a continuación se va a procesar la muestra $(\boldsymbol{x},c)=(( 4, 5)^t,2)$, cuántos vectores de pesos se modificarán?
1. 0
2. 2
3. 3
4. 4

## Solución


**Muestra de entrenamiento en notación homogénea:**

$$
\begin{align*}
  &\boldsymbol{x}_1 = (1, 4,5)^t; \quad y_1 = 2 \\
\end{align*}
$$

**Vectores de pesos:**

$$
\begin{align*}
  &\boldsymbol{w}_1 = (-2, -2, -6)^t \\
  &\boldsymbol{w}_2 = (-2, -2, -6)^t \\
  &\boldsymbol{w}_3 = (-2, -2, -4)^t \\
  &\boldsymbol{w}_4 = (-2, -2, -4)^t \\
\end{align*}
$$

**Evaluación de la condición $\forall c : c \neq 2$:**

- Clase incorrecta $1$ (causa error):

$$
\boldsymbol{w}_1^t\boldsymbol{x}_1+b>\boldsymbol{w}_2^t\boldsymbol{x}_1; \quad (-2, -2, -6) \cdot \begin{pmatrix} 1\\ 4\\ 5\end{pmatrix} + 0.1 > (-2, -2, -6) \cdot \begin{pmatrix} 1\\ 4\\ 5\end{pmatrix}; \quad -39.9 > -40 \quad \to \quad \text{True}
$$

- Clase incorrecta $3$ (causa error):

$$
\boldsymbol{w}_3^t\boldsymbol{x}_1+b>\boldsymbol{w}_2^t\boldsymbol{x}_1; \quad (-2, -2, -4) \cdot \begin{pmatrix} 1\\ 4\\ 5\end{pmatrix} + 0.1 > (-2, -2, -6) \cdot \begin{pmatrix} 1\\ 4\\ 5\end{pmatrix}; \quad -29.9 > -40 \quad \to \quad \text{True}
$$

- Clase incorrecta $4$ (causa error):

$$
\boldsymbol{w}_4^t\boldsymbol{x}_1+b>\boldsymbol{w}_2^t\boldsymbol{x}_1; \quad (-2, -2, -4) \cdot \begin{pmatrix} 1\\ 4\\ 5\end{pmatrix} + 0.1 > (-2, -2, -6) \cdot \begin{pmatrix} 1\\ 4\\ 5\end{pmatrix}; \quad -29.9 > -40 \quad \to \quad \text{True}
$$


**Respuesta correcta:** Se actualizarán los pesos de las tres clases incorrectas ($\boldsymbol{w}_1, \boldsymbol{w}_3, \boldsymbol{w}_4$) más los pesos de la clase correcta ($\boldsymbol{w}_2$). Por tanto, la respuesta correcta es la 4.

 En la tabla de la izquierda se proporciona un conjunto de $3$ muestras bidimensionales de aprendizaje de $3$ clases, mientras que en la tabla de la derecha se proporciona un conjunto de pesos iniciales por cada clase.
$$\begin{array}{ccc}
\begin{array}{c|ccc}

w_{c2} & -2 &  0 &  4\\
\end{array}

Se pide:
1. ($1.5$ puntos) Realiza una traza de ejecución de una iteración del algoritmo Perceptrón, con factor de aprendizaje $\alpha=1$, margen $\gamma=0.1$ y utilizando los pesos iniciales proporcionados.
2. ($0.5$ puntos) Representa las regiones de decisión del clasificador resultante, así como las fronteras de decisión necesarias para su representación.

## Solución

**Pesos tras una iteración de Perceptrón (1 muestra mal clasificada):**

$$\begin{array}{c|ccc}
 & \boldsymbol{w}_1 & \boldsymbol{w}_2 & \boldsymbol{w}_3\\\hline
w_{c0} & -1 & 0 & -2\\
w_{c1} & -2 & 0 & 4\\
w_{c2} & -2 & 0 & 4\\
\end{array}$$

**Fronteras de decisión:**

$$\begin{align*}
&g_1(\boldsymbol{x}) = g_2(\boldsymbol{x}) \;\Rightarrow\; \boldsymbol{w}_1^t\boldsymbol{x}=\boldsymbol{w}_2^t\boldsymbol{x}%
\;\Rightarrow\; -1-2x_1-2x_2=0 \;\Rightarrow\; x_2 = -x_1 -0.5\\
&g_1(\boldsymbol{x}) = g_3(\boldsymbol{x}) \;\Rightarrow\; \boldsymbol{w}_1^t\boldsymbol{x}=\boldsymbol{w}_3^t\boldsymbol{x}%
\;\Rightarrow\; -1-2x_1-2x_2=-2+4x_1+4x_2 \;\Rightarrow\; x_2 = -x_1 + \frac{1}{6}\\
&g_2(\boldsymbol{x}) = g_3(\boldsymbol{x}) \;\Rightarrow\; \boldsymbol{w}_2^t\boldsymbol{x}=\boldsymbol{w}_3^t\boldsymbol{x}%
\;\Rightarrow\; 0=-2+4x_1+4x_2 \;\Rightarrow\; x_2 = -x_1 + 0.5 
\end{align*}$$

**Representación gráfica de las tres regiones de decisión con las dos fronteras de decisión involucradas:**

# Ejercicio 4

Supón que estamos aplicando el algoritmo Perceptrón, con factor de aprendizaje $\alpha=1$ y margen $b=0.1$, a un conjunto de $3$ muestras bidimensionales de aprendizaje para un problema de $2$ clases. Se sabe que, después de procesar las primeras $2$ muestras, se han obtenido los vectores de pesos $ \boldsymbol{w}_1 =(0,0,1)^t\text{,}\; \boldsymbol{w}_2 =(0,0,-1)^t$. Así mismo, se sabe que, después de procesar la última muestra, $(\boldsymbol{x}_{3},c_{3})$, se obtienen los vectores de pesos $\boldsymbol{w}'_1 =(-1,-5,-2)^t,\,$ y $\,\boldsymbol{w}'_2 =(1,5,2)^t$. ¿Cuál de las siguientes muestras es esa última muestra?
1. $((2,4)^t,2)$
2. $((5,4)^t,2)$
3. $((5,3)^t,2)$
4. $((2,5)^t,2)$

## Solución

$(\boldsymbol{x}_{3},c_{3})$ debe ser $((5,3)^t, 2)$, puesto que verifica que:

$$
\begin{align*}
  \boldsymbol{w}'_1 &= \boldsymbol{w}_1 - \alpha \cdot \boldsymbol{x}_{3} = (0,0,1)^t - 1 \cdot (1,5,3)^t = (-1,-5,-2)^t\\
  \boldsymbol{w}'_2 &= \boldsymbol{w}_2 + \alpha \cdot \boldsymbol{x}_{3} = (0,0,-1)^t + 1 \cdot (1,5,3)^t = (1,5,2)^t\\
\end{align*}
$$

Por tanto, la respuesta correcta es la 3.
